### Handle missing data
The data guide states that all numeric missing values are denoted as -9999 and all the text missing values are denoted by "NA". Looking at df.info() will tell us the proportions of the missing values denoted as null, but this is not representative of the other missing valies denoted as -9999 or "NA". So lets use this information to look at the entire distribution of missing data and decide if we should drop rows/columns with those missing values, or use an imputation method.

Alaska and alberta dataframes are the only ones that have missing values denoted by null. Probe_coeff and soil_coeff don't have any nulls, but we will check all four dataframes for the other inconsistent/missing values.

```python
def check_missing_data(df, data_type):
    
    '''
    This function will check the dataframe for missing values based on the criteria explained in the sections 
    above (i.e., -9999, NA, or null)
    
    Args:
        df (Pandas Dataframe) : the dataframe we are checking for missing and inconsistent values
              data_type (str) : the datatype we will filter the df (since some missing values are denoted 
                                differently for numerical and text columns)
                                
    Returns: 
        dictionary : the proportions of TOTAL missing values (null and filled in with -9999 or "NA") of each column
    
    '''
    
    if data_type == "numerical":
        # get all columns with int data types
        cols = df.loc[:, (df.dtypes == "float64") | (df.dtypes == "int64")].columns 
        
        # count how many rows contain either a null or -9999 for numerical values 
        
        proportions = np.array(
            np.round(
                (((df[cols].isnull()) | (df[cols] == -9999)).sum()/len(df))*100, 
                    5
                )
            )
        return pprint.pformat(dict(zip(cols,  proportions)))
    
    elif data_type == "string": # change to string later
        cols = df.loc[:, (df.dtypes == "string")].columns 
        # count how many rows contain either a null or NA for text values 

        proportions = np.array(
            np.round(
                (((df[cols].isnull()) | (df[cols] == "NA")).sum()/len(df))*100, 
                    5
                )
            )
        return pprint.pformat(dict(zip(cols,  proportions)))
```

```python
# numerical only dataframes

print("% of missing data in alaska numerical columns:\n", check_missing_data(alaska, "numerical"), '\n')
print("% of missing data in alberta numerical columns:\n", check_missing_data(alberta, "numerical"), '\n')

# both numerical and strings

print("% of missing data in probe coefficient numerical columns:\n", check_missing_data(probe_coeff, "numerical"), '\n')
print("% of missing data in probe coefficient string columns:\n", check_missing_data(probe_coeff, "string"), '\n')

print("% of missing data in soil coefficient numerical columns:\n", check_missing_data(soil_coeff, "numerical"), '\n')
print("% of missing data in soil coefficient string columns:\n", check_missing_data(soil_coeff, "string"), '\n')
```

```python
# location should be a mid point between alaska and alberta

def find_midpoint(x1, y1, x2, y2):
    '''
    These values are a mean of the latitudes and longitudes. Will use this to center the folium map at the midpoint
    of all the latitudes and longitudes
    
    Arguments:
        x1 (float): mean of the alaska latitudes
        y1 (float): mean of the alberta latitudes
        x2 (float): mean of the alaska longitudes
        y2 (float): mean of the alberta longitudes

    Returns: 
        the midpoint of the latitudes and longitudes in the format of:
            [mdpt of latitude values, mdpt of longitude values]

    '''
    
    return [(x1+y1)/2, (x2+y2)/2]

midpoint = find_midpoint(
    x1=np.mean(probe_coeff[probe_coeff['area']=='Alaska'].latitude),
    y1=np.mean(probe_coeff[probe_coeff['area']=='Alberta'].latitude),
    x2=np.mean(probe_coeff[probe_coeff['area']=='Alaska'].longitude),
    y2=np.mean(probe_coeff[probe_coeff['area']=='Alberta'].longitude)
)

all_map = folium.Map(location=midpoint, tiles="OpenTopoMap", zoom_start=4)

for lat, long in zip(probe_coeff.latitude, probe_coeff.longitude):

    folium.Marker(location=[lat, long]).add_to(all_map)
        
all_map
```

```python
def find_peaks_region(df):

    all_peaks = []
    for i in range(1,5):
        peaks, _ = find_peaks(df[f'probe{i}_cal'], height=0)
        if len(peaks) == 0:
            all_peaks.append(-1)
        else:
            all_peaks.append(max(peaks))
    
    return all_peaks
```

```python
def plot_avg_with_max(df, time_unit):
    '''
    Plots the monthly and yearly averages of soil moisture contents 
    
    Arguments:
         df (Pandas df) : this is the region specific dataframe
        time_unit (str) : string used to identify the time unit we are taking averages for (i.e., year, month, etc.)
    
    Returns:
        Plots all the probe specific yearly/monthly/etc. averages of soil moisture
    
    '''
    
    peaks = find_peaks_region(df)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    for num in range(1,5):

        ax.plot(df[time_unit], df[f'probe{num}_cal'], label=f'probe{num}_cal')
        if peaks[num-1] != -1:
            x = df[time_unit][peaks[num-1]]
            y = np.array(df[df[time_unit]==x][f'probe{num}_cal'])
            ax.plot(
                x, y,
                'o'
            )
            ax.annotate(f"({x}, {round(y[0], 3)})", (x, y))

    if time_unit == 'month':
        plt.xlabel("month")
        plt.ylabel("average calculated volumetric soil moisture, %")
        plt.title("Monthly averaged volumetric soil moisture")
        
    elif time_unit == 'year':
        plt.xlabel("year")
        plt.ylabel("average calculated volumetric soil moisture, %")
        plt.title("Yearly averaged volumetric soil moisture")
    plt.legend()
```

```python
def moisture_soil_avg_df(region_df, probe_df):
    
    '''
    Creates a pandas df that joins the region specific dataframe with the dataframe with the probe coefficient 
    and soil specific percentages.
    
    Arguments: 
    region_df (Pandas df) : region specific dataframe with volumetric soil content
     probe_df (Pandas df) : dataframe with information about probes and soil percentages 
     
    Returns: 
    df (Pandas df) : joined region and probe dataframes with averages probe specific columns (averages probe 1-4)
                     into one
    '''
    
    df = pd.merge(region_df, probe_df, on = ['probe_id', 'record'], how='inner')
    
    probe_period_cols = ['pa_us_avg_1', 'pa_us_avg_2', 'pa_us_avg_3', 'pa_us_avg_4']
    volumetric_cols = ['probe1_cal', 'probe2_cal', 'probe3_cal', 'probe4_cal']
    
    string_cols = list(df.select_dtypes(include='string').columns)
    extra_cols = ['A', 'B', 'C', 'latitude', 'longitude']
    
    # calc the averages of probe specific cols
    df['probe_period_avg'] = df[probe_period_cols].mean(axis=1)
    df['volume_avg'] = df[volumetric_cols].mean(axis=1)
    
    # drop the extra cols
    df = df.drop(np.array(probe_period_cols+volumetric_cols+string_cols+extra_cols), axis = 1)

    return df
```

```python
alaska_and_probe = moisture_soil_avg_df(alaska_cleaned, probe_coeff)
alberta_and_probe = moisture_soil_avg_df(alberta_cleaned, probe_coeff)
```

```python
# group by the probe id
alaska_moisture_soil_avgs = pd.DataFrame(alaska_and_probe.groupby('probe_id').mean())
alberta_moisture_soil_avgs = pd.DataFrame(alberta_and_probe.groupby('probe_id').mean())

alaska_soil_moisture_corrs = alaska_moisture_soil_avgs[
    pct_cols+['volume_avg']
].drop('burned_moss_pct', axis=1).corr(method='pearson')


alberta_soil_moisture_corrs = alberta_moisture_soil_avgs[
    pct_cols+['volume_avg']
].drop('burned_moss_pct', axis=1).corr(method='pearson')
```

```python
fig = plt.figure()
plt.figure(figsize=(30, 28))

plt.subplot(2, 2, 1)
sns.heatmap(alaska_soil_moisture_corrs)
plt.title('Alaska soil type and moisture correlations')

plt.subplot(2, 2, 2)
sns.heatmap(alberta_soil_moisture_corrs)
plt.title('Alberta soil type and moisture correlations')
```

- moderate/higher percentages of organic contact soil.

**Alberta**: Higher volumetric soil moisture is associated with

- very high percentages of live moss
